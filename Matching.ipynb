```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('Scraped_Data.csv')
df2 = pd.read_csv('RsID.csv')
```

```python
#df1.head()
```

```python
#df2.head()
```

```python
df=pd.merge(df1, df2, on='RsID')
#df.head()
```

```python
df.drop(columns=['position','Position','Address','Chr','final_summary'],axis=1,inplace=True)
# df.head()
```

```python
df.dropna(how='all',subset=['sum_1','sum_2','sum_3'],inplace=True)
# df
```

```python
df.dropna(how='all',subset=['geno1','geno2','geno3'],inplace=True)
# df
```

```python
df=df.reset_index(drop=True)
```

```python
pd.DataFrame(df).to_csv('Matching.csv')
```

```python
dd=pd.read_csv('Matching.csv')
```

```python
dd
```

```python
new=[]
n=dd.shape[0]
lis1=dd['SNP']==dd['geno1']
lis2=dd['SNP']==dd['geno2']
lis3=dd['SNP']==dd['geno3']
for i in range(n):
    if lis1[i]:
        new.append(dd['sum_1'][i])
    elif lis2[i]:
        new.append(dd['sum_2'][i])
    elif lis3[i]:
        new.append(dd['sum_3'][i])
    else:
        new.append(np.NaN)
dd['Final_conclusion']=new
dd
```

```python
dd.dropna(how='all',subset=['Final_conclusion'],inplace=True)
```

```python
dd.drop(columns=['sum_1','sum_2','sum_3'],axis=1,inplace=True)
```

```python
dd=dd.reset_index(drop=True)
```

```python
dd
```

```python
pd.DataFrame(dd).to_csv('Matched.csv')
```
